```python
import pandas as pd
import xarray as xr
```

```python
wind_power_data = pd.read_csv('windpower_task0_updated.csv')

#rename time series column and remove timezone
wind_power_data = wind_power_data.rename(columns= {'Unnamed: 0' : 'time'},)
dt= wind_power_data.time
wind_power_data.time = pd.to_datetime(wind_power_data.time).dt.tz_localize(None)

#convert 'time' column to index
wind_power_data = wind_power_data.set_index('time')
```

```python
df_pow_26_feb = wind_power_data.loc[ '2000-02-26 00:00:00': '2000-02-26 23:00:00']
```

```python
#2000-02-26 weather csv, extra columns: cluster, distance and region-- shared by Toni
df_weather = pd.read_csv('./2002_02_26_test_merge.csv')
df_weather = df_weather.set_index('time')
df_weather.index = pd.to_datetime(df_weather.index)
```

```python
df_weather_se4 = df_weather[(df_weather['region'] == 'SE4')]
df_temp = df_weather_se4.pivot_table(index='time', aggfunc='mean')
```

```python
df_se4_merge = pd.merge(df_temp, df_pow_26_feb['SE4'], on='time')
df_se4_merge = df_se4_merge.rename(columns={'SE4':'Power'})
```

```python
df_se4_merge.to_csv('2000_02_26_weather_power_merged.csv')
```
